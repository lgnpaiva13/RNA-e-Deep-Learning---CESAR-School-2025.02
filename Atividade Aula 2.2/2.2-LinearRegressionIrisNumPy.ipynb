# Regressão Linear com NumPy

## Objetivos deste experimento

Este exemplo utiliza regressão linear para estimar o comprimento das sépalas da íris a partir do seu comprimento das pétalas.
Utiliza-se programação totalmente NumPy com:
- o **cálculo do gradiente**,
- a **função de perda** e
- a atualização dos parâmetros são feitos de forma explícita.
- A rede é criada com uma camada como uma classe

## Importação dos pacotes

```python
%matplotlib inline
import matplotlib.pyplot as plt   # para plotting

from sklearn.datasets import load_iris  # para carregar dataset

import numpy as np  # processamento matricial

np.random.seed(1234)
```

## Dataset

### Leitura dos dados

```python
iris = load_iris()
data = iris.data[iris.target==1,::2]  # comprimento das sépalas e pétalas, indices 0 e 2

x_train = data[:,0:1]
y_train = data[:,1:2]

n_samples = x_train.shape[0]
print('x_train.shape:',x_train.shape, x_train.dtype)
print('y_train.shape:',y_train.shape, y_train.dtype)

print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

### Normalização dos dados

```python
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()
```

```python
plt.scatter(x_train,y_train)
plt.xlabel('Comprimento sepalas')
plt.ylabel('Comprimento petalas')
```

## Modelo da rede

### Reta de ajuste

A equação da reta no plano necessita de dois parâmetros, aqui denominados $w_0$ (*bias*) e inclinação $w_1$. Veja figura:

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/linhareta.png)

A reta de ajuste será dada por:

$$ \hat{y} = w_0 + w_1 x $$

onde
* $w_1$ é o coeficiente angular da reta e
* $w_0$ é a interseção do eixo vertical quando x é igual a zero, também denominado de *bias*.
* $x$ é a variável de entrada (comprimento das sépalas) e
* $\hat{y}$ é a predição (comprimento estimado das pétalas).

### Linear Regression Model

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinearNeuronio.png)

### Formulação matricial do neurônio

* $\mathbf{w} = [w_0 \ w_1]$ é vetor linha de parâmetros e
* $\mathbf{x} = [1 \ x_0] $ é vetor linha de uma amostra (com *bias*).

$$ \hat{y} = 1 w_0 + x_0 w_1 $$

$$ [\hat{y}] = [1 \ x_0] \begin{bmatrix} w_0 \\ w_1 \end{bmatrix} $$

$$ \mathbf{\hat{y}} = \mathbf{x} \mathbf{w^T} $$

### Fazendo o predict de várias amostras

$$ \begin{matrix}
\begin{bmatrix}
\hat{y}_0\\
\hat{y}_1\\
\\
\hat{y}_{M-1}
\end{bmatrix} &=& \begin{bmatrix}
1 & x_0 \\
1 & x_1\\
\\
1 & x_{M-1} \\
\end{bmatrix}
&
\begin{bmatrix}
w_0\\
w_1
\end{bmatrix} \\
(M,1) & & (M,2) & (2,1) \\
\end{matrix} $$

### Preenchendo x com coluna de 1's para bias

```python
x_train
```

```python
np.ones(shape=(n_samples,1))
```

```python
x_train_bias = np.hstack([np.ones(shape=(n_samples,1)), x_train])
```

```python
x_train_bias
```

### Classe com o modelo da rede

```python
class Net():
    def __init__(self, n_in, n_out):
        self.w = np.random.uniform(-0.1,0.1,(n_out,n_in)) #inicialização dos parâmetros

    def forward(self, x_bias):
        return x_bias.dot(self.w.T)
```

### Criando a rede, instanciando o objeto model

```python
model = Net(2,1) # duas entradas (1 + x0) e uma saída y_pred
```

```python
model.w
```

### Testando o predict da rede (mesmo sem treiná-la)

```python
xin = np.random.randn(4,2) # 4 amostras, 2 entradas (1 + x0)
ypred = model.forward(xin)
ypred
```

```python
ypred = model.forward(x_train_bias)
```

## Treinamento via gradiente descendente

Função de perda:
$$ J(\hat{y_i},y_i) = \frac{1}{M} \sum_{i=0}^{M-1} (\hat{y_i} - y_i)^2 $$

Gradiente:
$$  \mathbf{\nabla{J_w}} = \frac{2}{M}\mathbf{x^T}(\mathbf{x w^T} - \mathbf{y}) $$

Atualização dos parâmetros pelo gradiente descendente:
$$ \mathbf{w} = \mathbf{w} − \eta (\mathbf{\nabla J_w})^T $$

### Laço de treinamento

```python
num_epochs = 100
learning_rate = 0.5
for epoch in range(num_epochs):

    # forward - predict
    y_pred = model.forward(x_train_bias)

    #loss cálculo da função de perda
    loss = np.square(y_pred - y_train).mean()

    # cálculo do gradiente pelas derivadas parciais
    w_grad = (2.0/n_samples) * (x_train_bias.T).dot(x_train_bias.dot(model.w.T) - y_train)

    # gradiente descendente
    model.w = model.w - learning_rate * w_grad.T

    # verbose
    if (epoch+1) % 1 == 0:
        print('Epoch[{}/{}], loss: {:.6f}'
              .format(epoch+1, num_epochs, loss))
```

### Atividade 2 — Perda (MSE) com parâmetros iniciais (sem treinamento)

Antes de qualquer treinamento, instanciamos um novo modelo com pesos aleatórios
e calculamos a perda MSE para servir de referência.

```python
# Atividade 2: MSE com pesos inicializados aleatoriamente (sem treinar)
np.random.seed(1234)          # mesma seed para reprodutibilidade
model = Net(2, 1)

y_pred_init = model.forward(x_train_bias)
loss_inicial = np.square(y_pred_init - y_train).mean()

print(f'Pesos iniciais: {model.w}')
print(f'Perda (MSE) inicial (sem treinamento): {loss_inicial:.6f}')
```

### Atividade 1 — Treinamento via Gradiente Descendente com registro de perdas

O laço de treinamento é modificado para guardar o valor da perda ao final de cada
época. Ao término, plota-se a curva de aprendizado (perda × época).

```python
# Atividade 1: Laço de treinamento com registro da perda por época
num_epochs    = 100
learning_rate = 0.5
perdas        = []          # lista que guardará a perda de cada época

for epoch in range(num_epochs):

    # Forward – predição
    y_pred = model.forward(x_train_bias)

    # Cálculo da função de perda (MSE)
    loss = np.square(y_pred - y_train).mean()
    perdas.append(loss)        # ← guarda a perda da época atual

    # Gradiente da função de perda em relação aos pesos
    w_grad = (2.0 / n_samples) * (x_train_bias.T).dot(
                x_train_bias.dot(model.w.T) - y_train)

    # Atualização dos parâmetros (gradiente descendente)
    model.w = model.w - learning_rate * w_grad.T

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1:3d}/{num_epochs}]  loss: {loss:.6f}')

loss_gd = perdas[-1]          # perda final após o treinamento
print(f'\nPerda final (Gradiente Descendente): {loss_gd:.6f}')
```

```python
# Curva de aprendizado: perda em função da época
plt.figure(figsize=(8, 4))
plt.plot(range(1, num_epochs + 1), perdas, color='royalblue', marker='o',
         markersize=3, linewidth=1.5, label='Perda (MSE)')
plt.xlabel('Época')
plt.ylabel('Perda (MSE)')
plt.title('Curva de Aprendizado — Gradiente Descendente')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Atividade 3 — Solução ótima analítica e comparação com Gradiente Descendente

A solução analítica (mínimos quadrados) fornece os pesos ótimos **exatos**:

$$\mathbf{w}_{opt} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$$

Esses pesos são carregados no modelo e a perda MSE é calculada para comparação.

```python
# Atividade 3: Solução ótima analítica
x = x_train_bias
y = y_train
w_opt = (np.linalg.inv((x.T).dot(x)).dot(x.T)).dot(y)   # equação normal

# Carregar os pesos ótimos no modelo
model_opt = Net(2, 1)
model_opt.w = w_opt.T

y_pred_opt = model_opt.forward(x_train_bias)
loss_analitico = np.square(y_pred_opt - y_train).mean()

print(f'Pesos ótimos (analítico): {w_opt.T}')
print(f'Pesos treinados (GD):     {model.w}')
print()
print('─' * 55)
print(f'  Perda inicial  (sem treino):         {loss_inicial:.6f}')
print(f'  Perda final    (Grad. Descendente):  {loss_gd:.6f}')
print(f'  Perda ótima    (Solução Analítica):  {loss_analitico:.6f}')
print('─' * 55)
print(f'  Diferença GD vs Analítico:           {abs(loss_gd - loss_analitico):.2e}')
```

```python
# Comparação visual das três retas: inicial, treinada e ótima
x_line = np.linspace(0, 1, 100).reshape(-1, 1)
x_line_bias = np.hstack([np.ones((100, 1)), x_line])

y_line_init = x_line_bias.dot(np.array([[0.0], [0.0]]))   # pesos zero como ilustração
y_line_gd   = x_line_bias.dot(model.w.T)
y_line_opt  = x_line_bias.dot(model_opt.w.T)

plt.figure(figsize=(8, 5))
plt.scatter(x_train, y_train, color='gray', alpha=0.6, label='Dados reais', zorder=3)
plt.plot(x_line, y_line_gd,  color='royalblue', linewidth=2,
         label=f'Grad. Descendente  (MSE={loss_gd:.4f})')
plt.plot(x_line, y_line_opt, color='crimson',    linewidth=2, linestyle='--',
         label=f'Solução Analítica  (MSE={loss_analitico:.4f})')
plt.xlabel('Comprimento das sépalas (normalizado)')
plt.ylabel('Comprimento das pétalas (normalizado)')
plt.title('Comparação: Gradiente Descendente vs Solução Analítica')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
